## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Let's start with age fixiing - 18 + only eligible for loan
data = data[data['Age'] >= 18]  # 6% data losses, but that was of no use
````

````python
## Change the Data Types
cols = ['Month','Age','SSN','Num_Bank_Accounts','Num_Credit_Card','Num_of_Loan',
        'Num_of_Delayed_Payment','Num_Credit_Inquiries']

data[cols] = data[cols].astype(int)
````

````python
data.info()
````

````python
# Rename Monthly inhand salary to monthly salary
data.rename(columns={'Monthly_Inhand_Salary': 'Monthly_Salary'}, inplace=True)
````

````python
# Credit Age :
data['Credit_History_Years'] = data['Credit_History_Age'] / 12 # credit months into years
data['Credit_Start_Age'] = data['Age'] - data['Credit_History_Years'] # Credit Age calculation
````

````python
# Clean Type Of Loan Column
# Update the not specified and no data values in Type of column values with Nan
data['Type_of_Loan'].replace(['Not Specified','No Data'], np.nan, inplace=True)

# Clean and Find unique type of loan and count their frequency
df_clean = (data['Type_of_Loan'].str.replace('and ', '').str.split(',').explode().str.strip())

# Finding most frequent Loan
df_frequent_loan = df_clean.value_counts().sort_values(ascending=False)
print(df_frequent_loan)
````

````python
### Feature Engineering and Creating meaningful KPI'S
## Credit Risk Analysis
# A. Repayment Risk
data['Delay_Severity'] = data['Delay_from_due_date'] * data['Num_of_Delayed_Payment']   # Sevear delay check
data['Chronic_Delay'] = (data['Delay_from_due_date'] > 20).astype(int)                  # Always a Late payer
data['Min_Payment_Flag'] = (data['Payment_of_Min_Amount'] == 'Yes').astype(int)         # Minm- Payment done only
data['Payment_Risk'] = data['Delay_from_due_date'] + data['Num_of_Delayed_Payment']

# B. Financial Stress
data['DTI'] = data['Outstanding_Debt'] / data['Annual_Income']                          #Debt to Income Ratio -  risk indicator
data['EMI_Ratio'] = data['Total_EMI_per_month'] / data['Monthly_Salary']                # Monthly Stress
data['Disposable_Income'] = data['Monthly_Salary'] - data['Total_EMI_per_month']        # Net Disposable Income

# C. Credit Behaviour risk
data['Inquiry_Level'] = data['Num_Credit_Inquiries']                                    # Desparate for loan
data['Total_Accounts'] = data['Num_Bank_Accounts'] + data['Num_Credit_Card']            # overload of credit

## Relationship Analysis
data['Savings_Ratio'] = data['Monthly_Balance'] / data['Monthly_Salary']                # Income stability
data['Debt_per_Loan'] = data['Outstanding_Debt'] / data['Num_of_Loan']                  # Average loan burden
data.loc[data['Num_of_Loan'] == 0, 'Debt_per_Loan'] = np.nan                            # no loan = debt per loan = nan
data['Has_Loan'] = (data['Num_of_Loan'] > 0).astype(int)                                # Atleast one loan

## Segmentation
data['Income_Group'] = pd.qcut(data['Annual_Income'], 3, labels=['Low','Mid','High'])   # Different Income levels
data['Age_Group'] = pd.cut(data['Age'], bins=[18,30,45,60,100])                         # Different Age groups
data['Experience_Level'] = pd.qcut(data['Credit_History_Years'], 3, labels=['Low','Medium','High']) # Credit History
````

````python
# Loan Effect on credit score
data.groupby('Has_Loan')['Credit_Score'].value_counts(normalize=True)
````

````python
# Credit History effect on Credit Score
data.groupby('Experience_Level')['Credit_Score'].value_counts(normalize=True)
````

````python
credit_result = data.groupby('Credit_Score')[[
    'Delay_from_due_date',
    'Num_of_Delayed_Payment',
    'Outstanding_Debt',
    'Credit_Utilization_Ratio',
    'Annual_Income',
    'Num_Credit_Card',
    'Num_Bank_Accounts',
    'Credit_History_Age'
]].mean().sort_values('Delay_from_due_date', ascending=True)
````

### What all manipulations have you done and insights you found?

#### ✅ Data Cleaning & Preparation
Filtered dataset to include only customers aged 18 years and above, ensuring alignment with real-world loan eligibility criteria

Converted relevant variables (e.g., number of accounts, loans, and inquiries) into appropriate numeric formats for accurate computation

Renamed Monthly_Inhand_Salary to Monthly_Salary for clarity and consistency


#### ✅ Feature Transformation

Converted credit history from months to years by creating Credit_History_Years

Derived Credit_Start_Age to estimate when a customer began using credit

#### 🔹 Handling Missing and Complex Data

Replaced non-informative values such as “Not Specified” and “No Data” in the loan type column with null values

Cleaned multi-valued loan entries by splitting and standardizing them to identify individual loan types and their frequencies

🔹 Feature Engineering (KPI Creation)

🔴 Repayment Risk Indicators

Delay_Severity: captures combined impact of delay duration and frequency

Chronic_Delay: identifies customers with persistent late payments

Min_Payment_Flag: flags customers paying only the minimum amount

🔴 Financial Stress Indicators

DTI (Debt-to-Income Ratio): measures overall financial burden

EMI_Ratio: captures monthly repayment pressure

Disposable_Income: indicates remaining income after EMI obligations

🔴 Credit Behaviour Indicators

Inquiry_Level: reflects frequency of credit inquiries (credit demand)

Total_Accounts: captures total credit exposure

🔵 Relationship Metrics

Savings_Ratio: measures financial stability

Debt_per_Loan: evaluates average loan burden per active loan

Has_Loan: distinguishes customers with and without active loans

🟢 Segmentation Features

Income_Group: categorizes customers into income segments

Age_Group: groups customers based on life stages

Experience_Level: segments customers based on credit history length
